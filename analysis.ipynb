```python
import pandas as pd
import matplotlib.pyplot as plt

%reload_ext nb_black
```

```python
DESK_STATUS = "sit"

df = pd.read_csv(f"{DESK_STATUS}_samples.csv")
```

```python
fig, axes = plt.subplots(6, 6, figsize=(15, 15), subplot_kw=dict(projection="3d"))

df_good = df[df["ps"] == 1.0]
df_bad = df[df["ps"] == 0.0]

NAMES = [
    "nose, 0",
    "left_eye_inner, 1",
    "left_eye, 2",
    "left_eye_outer, 3",
    "right_eye_inner, 4",
    "right_eye, 5",
    "right_eye_outer, 6",
    "left_ear, 7",
    "right_ear, 8",
    "mouth_left, 9",
    "mouth_right, 10",
    "left_shoulder, 11",
    "right_shoulder, 12",
    "left_elbow, 13",
    "right_elbow, 14",
    "left_wrist, 15",
    "right_wrist, 16",
    "left_pinky, 17",
    "right_pinky, 18",
    "left_index, 19",
    "right_index, 20",
    "left_thumb, 21",
    "right_thumb, 22",
    "left_hip, 23",
    "right_hip, 24",
    "left_knee, 25",
    "right_knee, 26",
    "left_ankle, 27",
    "right_ankle, 28",
    "left_heel, 29",
    "right_heel, 30",
    "left_foot_index, 31",
    "right_foot_index, 32",
]

for i in range(33):
    ax = axes[i // 6][i % 6]

    ax.scatter(
        df_good[f"s0l{i}x"],
        df_good[f"s0l{i}y"],
        df_good[f"s0l{i}z"],
        c="green",
        label="good",
    )
    ax.scatter(
        df_bad[f"s0l{i}x"], df_bad[f"s0l{i}y"], df_bad[f"s0l{i}z"], c="red", label="bad"
    )
    ax.set_title(f"{names[i]}")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.legend()

fig.delaxes(axes[5][3])
fig.delaxes(axes[5][4])
fig.delaxes(axes[5][5])

plt.tight_layout()
```
