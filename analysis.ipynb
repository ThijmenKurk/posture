```python
import pandas as pd
import matplotlib.pyplot as plt

%reload_ext nb_black
```

```python
DESK_STATUS = "sit"

df = pd.read_csv(f"{DESK_STATUS}_samples.csv")
```

```python
fig, axes = plt.subplots(6, 6, figsize=(15, 15), subplot_kw=dict(projection="3d"))

df_good = df[df["ps"] == 1.0]
df_bad = df[df["ps"] == 0.0]

for i in range(33):
    ax = axes[i // 6][i % 6]

    ax.scatter(
        df_good[f"s0l{i}x"],
        df_good[f"s0l{i}y"],
        df_good[f"s0l{i}z"],
        c="green",
        label="good",
    )
    ax.scatter(
        df_bad[f"s0l{i}x"], df_bad[f"s0l{i}y"], df_bad[f"s0l{i}z"], c="red", label="bad"
    )
    ax.set_title(f"l{i}")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.legend()

fig.delaxes(axes[5][3])
fig.delaxes(axes[5][4])
fig.delaxes(axes[5][5])

plt.tight_layout()
```
